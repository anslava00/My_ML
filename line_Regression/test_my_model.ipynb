```python
y = df.loc[:, 'medv']
```

```python
X = df.loc[:, ['rm']]
```

```python
liner_model.fit(X, y)
```

```python
list(liner_model.w)
```

```python
plt.scatter(x=X, y=y)
liner_line = [[i] for i in range(3, 10)]
plt.plot(liner_line, liner_model.predict(liner_line), color='green')
```
